# Diabetes Prediction Dataset

Welcome to the Diabetes Prediction Dataset, a valuable resource for researchers, data scientists, and medical professionals interested in the field of diabetes risk assessment and prediction. This dataset contains a diverse range of health-related attributes, meticulously collected to aid in the development of predictive models for identifying individuals at risk of diabetes. By this dataset, we aim to foster collaboration and innovation within the data science community, leading to improved early diagnosis and personalized treatment strategies for diabetes.

- Id: Unique identifier for each data entry.
- Pregnancies: Number of times pregnant.
- Glucose: Plasma glucose concentration over 2 hours in an oral glucose tolerance test.
- BloodPressure: Diastolic blood pressure (mm Hg).
- SkinThickness: Triceps skinfold thickness (mm).
- Insulin: 2-Hour serum insulin (mu U/ml).
- BMI: Body mass index (weight in kg / height in m^2).
- DiabetesPedigreeFunction: Diabetes pedigree function, a genetic score of diabetes.
- Age: Age in years.

#### Model Building

```python
#Select the features
X = healthcare_data.drop('Outcome',axis=1)
Y = healthcare_data[['Outcome']]
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV,StratifiedKFold
```

```python
#Split the training and testing set
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42,stratify=Y)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
```

```python
X_train_res,Y_train_res = smote.fit_resample(X_train,Y_train)
```

```python
from collections import Counter
```

```python
Counter(Y_train)
```

```python
Counter(Y_train_res)
```

```python
from sklearn.linear_model import LogisticRegression

lgr = LogisticRegression()
```

```python
cv = StratifiedKFold(n_splits=4,shuffle=True)
```

```python
param_grid = {
    'penalty': ['l1', 'l2'],              
    'C': [0.01, 0.1, 1, 10, 100],         
    'solver': ['liblinear', 'saga'],      
    'class_weight': ['balanced']    
}
```

### Logistic Regression

```python
lgr_model = GridSearchCV(estimator=lgr,param_grid=param_grid,cv=cv,verbose=2,scoring='accuracy')
```

```python
lgr_model.fit(X_train_res,Y_train_res)
```

```python
best_lgr_model = lgr_model.best_estimator_
```

```python
lgr_model.best_params_
```

```python
lgr_model.best_score_
```

```python
lgr_model.score(X_test,Y_test)
```

```python
lgr_model.score(X_train_res,Y_train_res)
```

```python
yhat = lgr_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,classification_report,confusion_matrix
```

```python
accuracy_score(Y_test,yhat)
```

```python
precision_score(Y_test,yhat)
```

```python
recall_score(Y_test,yhat)
```

```python
print(classification_report(Y_test,yhat))
```

```python
plt.figure(figsize=(8,4))
sns.heatmap(confusion_matrix(Y_test,yhat),annot=True,cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
feature_importance_df = pd.DataFrame(
    {
        "Features":X.columns,
        "Feature Importance":best_lgr_model.coef_[0]
    }
).sort_values(by='Feature Importance',ascending=False)
```

```python
X = healthcare_data[['Glucose','BMI','Pregnancies','DiabetesPedigreeFunction','Age']]
Y = healthcare_data['Outcome']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=42,stratify=Y)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],   
    'max_depth': [5, 10, 15, 20],     
    'min_samples_split': [2, 5, 10],        
    'min_samples_leaf': [1, 2, 4],         
    'class_weight': ['balanced']
}
```

```python
model_rf = GridSearchCV(estimator=rf,param_grid=param_grid,cv=cv,scoring='accuracy',verbose=2)
```

```python
X_train_res,Y_train_res = smote.fit_resample(X_train,Y_train)
```

```python
model_rf.fit(X_train_res,Y_train_res)
```

```python
model_rf.score(X_test,Y_test)
```

```python
model_rf.score(X_train_res,Y_train_res)
```

```python
yhat = model_rf.predict(X_test)
```

```python
accuracy_score(Y_test,yhat)
```

```python
precision_score(Y_test,yhat)
```

```python
recall_score(Y_test,yhat)
```

```python
f1_score(Y_test,yhat)
```

```python
print(classification_report(Y_test,yhat))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(model_rf,X,Y,cv=5,scoring='accuracy')
```

```python
scores.mean()
```

```python
plt.figure(figsize=(8,4))
sns.heatmap(confusion_matrix(Y_test,yhat),annot=True,fmt='d',cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

##### Key Observations
- The accuracy of the model is 98% i.e 98% of model's prediction are correct
- 98% of data is correctly predicted as Non-Diabetic
- 97% of data is correctly predicted as Diabetic
- 99% of data of actual Non-Diabetic patients are correctly predicted
- 97% of data of actual Diabetic patients are correctly predicted

```python
best_model = model_rf.best_estimator_
```

```python
feature_imp_df = pd.DataFrame({
    'Feature':X.columns,
    'Feature Importance':best_model.feature_importances_
}).sort_values(by='Feature Importance',ascending=False)
```

```python
feature_imp_df
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Feature',y='Feature Importance',data=feature_imp_df)
plt.title('Feature Importance Graph')
plt.xlabel('Features')
plt.ylabel('Feature Importance')
plt.show()
```

##### Key Observations
1. Glucose dominates  makes sense medically; high glucose is a strong indicator of diabetes.
2. BMI is also critical  aligns with obesity-related diabetes risk.
3. Age and DiabetesPedigreeFunction  moderate contributors, still valuable.
4. Pregnancies  contributes least among top 5, but still non-negligible.
- Skipping other features (Insulin, SkinThickness, BloodPressure) was a good decision  they had very low or negative importance before.
